```python
data.plot(figsize=(12, 6))
plt.title('Unemployment Rate over Time')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
column_name = ' Estimated Unemployment Rate (%)' 
average_unemployment = data[column_name].mean()
max_unemployment = data[column_name].max()
min_unemployment = data[column_name].min()
```
